```python
import cv2
```

```python
from deepface import DeepFace
```

```python
img = cv2.imread('happy-boy-smile.jpg')
img2 = cv2.imread('sad-boy-.jpg')
img3 = cv2.imread('Angry-face.jpg')
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(img)
```

```python
plt.imshow(img2)
```

```python
plt.imshow(img3)
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
plt.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
```

```python
plt.imshow(cv2.cvtColor(img3, cv2.COLOR_BGR2RGB))
```

```python
# Analyze the image for emotions
predictions = DeepFace.analyze(img, actions=['emotion'], enforce_detection=False)

# Format and print the predictions in a vertical list
for key, value in predictions[0].items():
    if isinstance(value, dict):  # Handle nested dictionaries
        print(f"{key}:")
        for sub_key, sub_value in value.items():
            print(f"  {sub_key}: {sub_value}")
    else:
        print(f"{key}: {value}")
```
